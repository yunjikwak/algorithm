# 리스트 회전

```python
# 시계방향 90도 회전 (= 반시계 방향 270)
def rotate_arr_90(arr):
    rotated_arr = []
    # 배열을 상하 반전시킨 후 -> 행열 바꾸기
        # arr[::-1] <- 행 순서 뒤집기(상하 반전)
        # *arr 로 개별 인자로 풀고(즉 열 list로 남김) 
        # zip <- 각 인자의 같은 인덱스에 있는 요소끼리 묶어줌
    for row in zip(*arr[::-1]):
        rotated_arr.append(list(row))
    
    return rotated_arr
```

1. arr[::-1]: 배열 arr의 행 순서를 거꾸로 뒤집습니다 (상하 반전).
- 예시: `[[1, 2], [3, 4]]` -> `[[3, 4], [1, 2]]`
2. *arr[::-1]: * 연산자는 리스트의 각 요소를 개별 인자로 "풀어"줍니다.
- 예시: `[[3, 4], [1, 2]]` -> `[3, 4]` 와 `[1, 2]` 라는 두 개의 인자로 분리됩니다.
3. zip([3, 4], [1, 2]): zip 함수는 각 인자의 같은 인덱스에 있는 요소끼리 묶어줍니다.
- 첫 번째 요소들: 3과 1이 묶여 `(3, 1)` 튜플 생성
- 두 번째 요소들: 4와 2가 묶여 `(4, 2)` 튜플 생성
- 결과적으로 zip은 `(3, 1)`과 `(4, 2)`를 차례로 반환하는 이터레이터(iterator)가 됩니다. 이는 행과 열을 바꾸는 **전치(Transpose)**와 같은 효과입니다.
4. list(row): zip이 반환한 튜플 `(3, 1)`을 리스트 `[3, 1]`로 변환하여 rotated_arr에 추가합니다.

```python
# 시계방향 180도 회전 (= 반시계 방향 180)
def rotate_arr_180(arr):
    rotated_arr = []
    # 상하 반전 -> 행 좌우 반전
    for row in arr[::-1]: # 상하 반전
        rotated_arr.append(row[::-1]) # 행의 요소(열) 뒤집기 -> 좌우 반전
    
    return rotated_arr
```

```python
# 시계방향 270도 회전 (= 반시계 방향 90)
def rotate_arr_270(arr):
    rotated_arr = []
    # 배열 행 <-> 열 변경 -> 전체 상하 반전
        # zip(*arr) : 배열 행열 바꾸기(전치)
        # 전체 -> list로 만들기
        # [::-1] : 리스트 순서 뒤집기 -> 상하 반전
    for row in list(zip(*arr))[::-1]:
        rotated_arr.append(list(row)) # 각 행 정보 -> 튜플에서 리스트로
    
    return rotated_arr
```

동작 원리:

1. `zip(*arr)`: 먼저 원래 배열의 행과 열을 바꿉니다 (전치).
    - 예시: `[[1, 2], [3, 4]]` -> `zip` 결과는 `(1, 3)`, `(2, 4)`
2. `list(zip(*arr))`: `zip` 결과를 리스트로 만듭니다.
    - 예시: `[(1, 3), (2, 4)]`
3. `[::-1]`: 이 리스트의 순서를 거꾸로 뒤집습니다 (상하 반전).
    - 예시: `[(2, 4), (1, 3)]`
4. `list(row)`: 각 튜플을 리스트로 변환하여 `rotated_arr`에 추가합니다.

```python
# transpose / 전치 행렬
def transpose_arr(arr):
    transposed_arr = []
    for row in zip(*arr): # 행 열 전치
        transposed_arr.append(list(row))
        
    return transposed_arr
```

`zip(*arr)`는 파이썬에서 2차원 배열의 행과 열을 바꾸는 대표적인 방법(idiom)입니다.

1. `arr`: 배열의 각 행 리스트를 개별 인자로 풀어줍니다.
    - 예시: `[[1, 2, 3], [4, 5, 6]]` -> `[1, 2, 3]` 과 `[4, 5, 6]`
2. `zip([1, 2, 3], [4, 5, 6])`: `zip`이 각 리스트의 같은 인덱스 요소를 묶어줍니다.
    - `1`과 `4`가 묶여 `(1, 4)`
    - `2`와 `5`가 묶여 `(2, 5)`
    - `3`과 `6`이 묶여 `(3, 6)`
3. 결과적으로 `(1, 4)`, `(2, 5)`, `(3, 6)`이 새로운 행이 되어 원래 배열의 행과 열이 바뀐 효과를 냅니다.

# 부분 회전

```python
arr = [[7 * j + i for i in range(1, 8)] for j in range(7)]
new_arr = [[0] * 7 for i in range(7)]
sy, sx = 2, 2
length = 3
```

```python
# 배열 특정 부분(정사각형) 회전시키기
def rotate_90(sx, sy, length):
    global arr, new_arr
    
    for x in range(sx, sx+length):
        for y in range(sy, sy+length):
            # 1단계 : (0,0)으로 옮겨주는 변환 -> 상대 좌표 가지기
            ox, oy = x - sx, y - sy
            # 2단계 : 90도 회전했을 때의 좌표 구하기 -> 회전시킴
            rx, ry = oy, length - ox - 1
            # 3단계 : 다시 (sx, sy) 더해주기 -> 절대 좌표
            new_arr[sx+rx][sy+ry] = arr[x][y]
            
    # new_arr 값 ~> 현재 board에 옮기기
    for x in range(sx, sx+length):
        for y in range(sy, sy+length):
            arr[x][y] = new_arr[x][y]
```

# 순열, 중복 순열, 조합, 중복 조합

### 순열 (permutations)

```python
arr = [1,2,3,4]
visited = [0] * len(arr)

def permutations(n, new_arr):
    # 순서 상관 O, 중복 X
    if len(new_arr) == n:
        return 
    
    for i in range(len(arr)):
        if not visited[i]:
            visited[i] = 1
            permutations(n, new_arr+[arr[i]])
            visited[i] = 0
            
permutations(2, [])
```

- 순서 O, 중복 X
- visited 배열 사용(순서 고려하기 때문)
- `parameter` : 뽑을 개수, 뽑힌 값을 담는 배열

### 중복 순열(product)

```python
arr = [1, 2, 3, 4]

def product(n, new_arr):
    global arr
    
    # 순서 상관 0, 중복 0
    if len(new_arr) == n:
        return
    
    for i in range(len(arr)):
        product(n, new_arr + [arr[i]])

product(2, [])
```

- 순서 0, 중복 0
- 부가 조건이 없는 가장 라이트한 코드라고 생각하기
- `parameter` : 뽑을 개수, 뽑힌 값을 담는 배열

### 조합(combinations)

```python
arr = [1, 2, 3, 4]

# 현재 인덱스 +1 만큼을 매개변수로 계속 넘겨주어야함
    # 순서 X 중복 X => 현재 인덱스보다 같거나 작은 인덱스는 볼 필요 X

def combinations(n, new_arr, c):
    # 순서 상관 X, 중복 X
    if len(new_arr) == n:
        return
    
    for i in range(c, len(arr)):
        combinations(n, new_arr + [arr[i]], i + 1)

combinations(2, [], 0)
```

- 순서 X, 중복 X
- 현재 인덱스를 의미하는 매개변수가 추가됨
    - 순서가 상관없고 중복 불가이기 때문에 현재 인덱스보다 같거나 작은 인덱스는 볼 필요가 없기 때문임. 따라서 재귀돌릴때 `현재 인덱스+1` 값을 넘김
- `parameter` : 뽑을 개수, 뽑힌 값을 담는 배열, 현재 인덱스

### 중복 조합 (combinations_with_replacement)

```python
arr = [1, 2, 3, 4]

def combinations_with_replacement(n, new_arr, c):
    # 순서 상관 X, 중복 O
    if len(new_arr) == n:
        return
    
    for i in range(c, len(arr)):
        combinations(n, new_arr + [arr[i]], i)


combinations_with_replacement(2, [], 0)
```

- 순서 X, 중복 0
- 조합과 동일하게 현재 인덱스를 의미하는 매개변수가 추가됨
    - 중복 가능하므로 재귀 돌릴때 현재 인덱스+1가 아닌 `현재 인덱스` 값을 그대로 넘김
- `parameter` : (뽑을 개수, 뽑힌 값을 담는 배열, 현재 인덱스)
